# Variation of people divided by gender, who work as full time employees and say that the type of work they do primarily involves manual and physical labor

```r
employees <- sqldf("SELECT * from pew_data where emplnw = 1 or emplnw = 2 or emplnw = 5")
employees <- within(employees , organization <- factor(emptype1, labels = c("Large corporation", "Medium-size company", "Small business", "Government", "Educational Institution" , "Non Profit organisation", "None", "None", "None")))
employees <- within(employees, sex <- factor(sex, labels = c("Male","Female")))
employees <- within(employees, type <- factor(emptype2, labels = c("Involves manual labor","Doesn't involve manual labor","Don't know")))
```

```r
qplot(sex, data=employees, geom="bar", facets = . ~ type, fill = organization)
```

### From the graph above, we can say that men are involved in more physical labor while both men and women are fairly distributed among the employees who are not involved in manual or physical labor
