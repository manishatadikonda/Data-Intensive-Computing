```r
install.packages("XML", repos='http://cran.us.r-project.org')
```

```r
library(XML)
```

```r
link <- 'http://aiweb.cs.washington.edu/research/projects/xmltk/xmldata/data/courses/reed.xml'
data <- xmlTreeParse(link,useInternalNode=TRUE)
```

```r
data
```

```r
dataframe <- xmlToDataFrame(data)
```

```r
print(dataframe)
```

```r
write.csv(dataframe, file="xmlToDF.csv")
```
